```python
import pandas as pd
import numpy as np

# ساخت دیتای نمونه حرفه‌ای‌تر
data = {
    "date": pd.date_range(start="2024-01-01", periods=30),
    "product": np.random.choice(["Laptop", "Mobile", "Tablet"], 30),
    "price": np.random.randint(500, 2000, 30),
    "quantity": np.random.randint(1, 5, 30)
}

df = pd.DataFrame(data)

# ساخت ستون فروش کل
df["total_sales"] = df["price"] * df["quantity"]

df.head()
```

```python
# فروش کل ماه
total_month_sales = df["total_sales"].sum()
print("Total Month Sales:", total_month_sales)

# فروش بر اساس محصول
sales_by_product = df.groupby("product")["total_sales"].sum()
print("\nSales by Product:")
print(sales_by_product)

# میانگین فروش روزانه
daily_average = df["total_sales"].mean()
print("\nDaily Average Sales:", daily_average)
```

```python
import matplotlib.pyplot as plt

sales_by_product.plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
# پیدا کردن روز با بیشترین فروش
best_day = df.loc[df["total_sales"].idxmax()]
print("Best Sales Day:")
print(best_day)
```

```python
df.to_excel("sales_analysis.xlsx", index=False)
```

```python
average_price = df.groupby("product")["price"].mean()
print(average_price)
```

```python
# محاسبه فروش روزانه
daily_sales = df.groupby("date")["total_sales"].sum()

# رسم نمودار روند فروش
daily_sales.plot(kind="line")
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
# محاسبه مجموع فروش هر محصول
product_sales = df.groupby("product")["total_sales"].sum()

# محاسبه کل فروش ماه
total_month_sales = product_sales.sum()

# محاسبه درصد سهم هر محصول
sales_percentage = (product_sales / total_month_sales) * 100

sales_percentage
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.pie(product_sales, labels=product_sales.index, autopct='%1.1f%%')
plt.title("Sales Percentage by Product")
plt.show()
```
